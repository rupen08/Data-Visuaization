## bar graph

```python
## good use of bar chart is specially when we want to see the value counts mostly which are independent from each other
## also good for series operation
```

```python
def performance(stock_price):
    if stock_price <= 200:
        return "Poor"
    elif stock_price > 200 and stock_price <= 500:
        return "Good"
    else:
        return "Excellent"
```

```python
bb["Open"].apply(performance).value_counts().plot()

## this is just plot method and as we can see all 3 parameter make no sense as they are independent
```

## bar pot

```python
bb["Open"].apply(performance).value_counts().plot(kind = "bar")
plt.ylabel("Value counts for the condition")
## now its bar
## x axis is the value counts for 3 categories not the $ 
```

```python
bb["Open"].apply(performance).value_counts()
## means there are 1864 values that has stock values > 500$ and so on 
```

```python
bb["Open"].apply(performance).value_counts().plot(kind = "barh")
```

## pie chart

```python
bb["Open"].apply(performance).value_counts().plot(kind = "pie")
```

## histogram

```python
## it is used basically to find out the distribution or finding out how thw data is spread
```

```python
## let say i want see how the stock price is between 0-100 $ gap
```

```python
def custom_round(stock_price):
    round_fig = int (stock_price / 100.00) * 100
    return round_fig
    
```

```python
bb["Open"].apply(custom_round).value_counts().sort_index()
```

```python
bb["Open"].apply(custom_round).nunique()
```

```python
bb["Open"].apply(custom_round).value_counts().sort_index().plot(kind = "hist")
```

```python
## see the gap
bb["Open"].apply(custom_round).value_counts().sort_index().plot(kind = "hist", bins = 21)
```
