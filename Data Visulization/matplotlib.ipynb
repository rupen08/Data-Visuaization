```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
gas = pd.read_csv("gas_prices.csv")
```

```python
gas.head(3)
```

## Bar Plot

```python
plt.figure(figsize= (10,6))

for country in gas:
    if country != "Year":
        plt.bar(country, gas[country].mean())
```

## line plot

```python
plt.figure(figsize = (8,5))

## always incluse this at the front

plt.plot(gas["Year"],gas["Japan"], label = "Japan", marker = ".")
plt.plot(gas["Year"],gas["Mexico"], label = "Mexico", marker = ".")
plt.plot(gas["Year"],gas["Canada"], label = "Canada", marker = ".")

plt.title("Gas prices for different countries (USD)", fontdict = { "fontweight" : "bold", "fontsize" : 14})

## the label thing works here
plt.legend()
plt.xlabel("Years")
plt.ylabel("USD")

## but years are not perfect
## too close x  axis so xticks makes it clear but gives you all values that is why it was tight
## this helps to look good

plt.xticks(gas["Year"][::2])

## if want to save, do it before show

plt.savefig("gas_price.png", dpi = 300)
           
plt.show()
```

```python
plt.figure(figsize = (8,7))
plt.title("Gas prices for different countries (USD)", fontdict = { "fontweight" : "bold", "fontsize" : 20})
plt.xlabel("Years")
plt.ylabel("USD")
## the label thing works here

## but yearsa re not perfect

plt.xticks(gas["Year"][::2])
## too close x  axis so xticks makes it clear but gives you all values that is why it was tight
## this helps to look good
for country in gas:
    if country != "Year":
        plt.plot(gas["Year"], gas[country], label = country, marker = ".")
        ## use the same code line as previous
plt.legend(loc = 2)
## loc term helps to set the position of the legend
plt.show()
```

## pie charts

```python
left = fifa[fifa["Preferred Foot"] == "Left"].count()[0]
right = fifa[fifa["Preferred Foot"] == "Right"].count()[0]

plt.pie([left, right], shadow = True, autopct='%1.1f%%', labels = ["left", "right"] )
plt.title("Foot Preferred")
## have to pass on a list

plt.show()
```

```python
fifa.shape
```

```python
def foot(foot_name):
    if foot_name == "Left":
        return "Left"
    else:
        return "Right"
        
fifa["Preferred Foot"].apply(foot).value_counts().plot(kind = "pie", shadow = True, autopct='%1.1f%%')

plt.show()

## second  way
```
