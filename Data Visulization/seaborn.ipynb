```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## loading some internal dataset from seaborn
```

```python
car = sns.load_dataset("car_crashes")
car.head(3)
```

## distribution plot

```python
sns.distplot(a = car["not_distracted"], kde = True)
## its histogram with kernal density estimation (kde = true) by defualt
plt.show()
```

## joint plot

```python
## use to compare two distribution 
## scatter plot by defualt
```

```python
sns.jointplot(x = "speeding", y = "alcohol", data = car, kind = "kde")

## can change the kind to what we want
```

 ## kde plot

```python
sns.kdeplot(car["alcohol"])
```

## pair plot

```python
sns.pairplot(car)
## relationship across the dataframe
```

```python
## onr more dataframe
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head(3)
```

```python
## plot a pairplot for this df
```

```python
sns.pairplot(tips, hue = "sex", palette = "bright")
## hue works when we pass categorical data and palette gives wanted color
## deep , muted , pastel , bright , dark , and colorblind for palette choise
```

## rug plot

```python
sns.rugplot(tips["total_bill"])
```

## styling

```python
sns.set_style("darkgrid")
## style must be one of white, dark, whitegrid, darkgrid, ticks
sns.set_context ("paper", font_scale = 1.6)
## context must be in paper, notebook, talk, poster

plt.figure(figsize = (8,7))

sns.jointplot(x = "tip", y = "size", data = tips, kind = "reg")
```

## categorical plotting

## 1. bar plot

```python
## distribution using category data with reference of one numerical data
```

```python
mask = tips["sex"] == "Female"
tips[mask]["total_bill"].median()
```

```python
import numpy as np
sns.barplot(x = "sex", y = "total_bill", data = tips, estimator = np.mean, ci = 80, capsize = .2)
## the data is plot as a mean estimator by defualt but we can change it
## here the black line show the confidence intervel, we can set ci = any number, but defualt is 95%
## capsize gives horizontal cap on top and bottom of the line
```

## 2. count plot

```python
sns.countplot(x = "sex", data = tips)
```

## 3. box plot

```python
## compare different variables and will show quartile of data
```

```python
plt.figure(figsize = (8,7))
sns.boxplot(x = "day", y = "total_bill", data = tips, hue = "sex")
plt.legend()
```

## violin plot

```python
## combination of box and kde plot
```

```python
plt.figure(figsize = (8,7))
sns.violinplot(x = "day", y = "total_bill", data = tips, hue = "sex")
```

## strip plot

```python
plt.figure(figsize = (8,7))
sns.set_style(style = "white")
sns.stripplot(x = "day", y = "total_bill", data = tips, hue = "sex", dodge = True)
```

## swarm plot

```python
sns.swarmplot(x = "day", y = "total_bill", data = tips)
```

## heatmaps

```python
plt.figure(figsize = (8,7))

sns.set_context(context = "paper", font_scale = 1.5)
car_crash = car.corr()
car_crash
sns.heatmap(data = car_crash, annot = True)

## annot just gives values in the box
```
